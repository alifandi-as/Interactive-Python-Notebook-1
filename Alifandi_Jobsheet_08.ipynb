# **Praktikum 04 - Fungsi dengan args dan kwargs**

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# **Praktikum 07 - Fungsi Decorator**

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

# **Penugasan 6: Scope Variabel**

```python
x = 100 // Ini merupakan variabel global

def ubah():
  // Ketika variabel didefinisikan di dalam fungsi, itu menjadi variabel lokal
  // Variabel lokal hanya berlaku dalam fungsi di mana itu didefinisikan
  // Sedangkan variabel global dapat digunakan di manapun selama tidak ada variabel lokal yang sama
  x = 50
  print(f"x lokal: {x}")

// Variabel lokal tidak memengaruhi variabel global yang sama
ubah()
print(f"x global: {x}")
```

# **Penugasan 8: Fungsi Decorator**

```python
def logger(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan")
  return wrapper

@logger
def halo():
  print("Hello world!")

halo()
```

# **Penugasan 9: Studi Kasus Nyata - Program Nilai**

```python
def input_nilai():
  nilai = []
  while True:
    masukan = input("Masukkan nilai (ketik q untuk berhenti): ")

    if masukan == "q":
      break
    nilai.append(int(masukan))
  return nilai

def hitung_rata(nilai):
  print(f"Sum {sum(nilai)} Len {len(nilai)}")
  return sum(nilai) / len(nilai)

def status_kelulusan(nilai):
  if (nilai >= 70):
    print("Anda lulus")
  else:
    print("Anda tidak lulus")

nilai = input_nilai()
rata_rata = hitung_rata(nilai)
status_kelulusan(rata_rata)

```
