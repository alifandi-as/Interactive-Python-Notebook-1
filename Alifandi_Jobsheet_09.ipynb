# **Penugasan 6: Studi Kasus Mini**

```python
import random, laporan

hasil = []

for i in range(10):
  a = random.randint(1,10)
  b = random.randint(1,10)
  res = int(input(f"{a} + {b} = "))

  hasil.append(res == a + b)
  if (res == a + b):
    print("Jawaban Anda benar")
  else:
    print("Jawaban Anda salah")

laporan.lapor(hasil)
```
